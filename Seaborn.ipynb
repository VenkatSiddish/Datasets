# KDE Plot: Kernel Density Estimation

It allows to estimate probability density function from a certain data set by non-parametrica way

 UNIVIRATE (1D)

```python
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower, shade=True);
```

```python
sns.kdeplot(cars.horsepower, bw=10);
```

```python
sns.kdeplot(cars.horsepower, cumulative=True);  #switch over from probability df to cumulative df
```

BIVARIATE 2D

```python
sns.kdeplot(cars.horsepower, cars.mpg);
```

# Violin plot

```python
# 1 boxplot and 2 kdeplots (mirror) gives a violin plot
# catageorical plot similar to boxplot median can be known
```

```python
cars=sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars=sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement)
```

```python
sns.violinplot(x='cylinders', y='displacement',hue='origin',data=cars)
```

```python
sns.violinplot(x='cylinders', y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])])
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])], split=True)
plt.legend(loc=2)
```

```python
sns.violinplot(x='cylinders', y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])], split=True, 
               inner='quartiles')
#shows 25,75 percentile of data and also median
```

```python
sns.violinplot(x='cylinders', y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])], split=True,
              scale='count')
#count shows no.of data points
```

```python
sns.violinplot(x='cylinders', y='displacement',hue='origin',data=cars[cars.origin.isin(['japan','europe'])], split=True,
              scale='count',scale_hue=False)
plt.legend(loc=2)
```
